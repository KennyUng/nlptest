```python
# Setting up training data
training_data = []

# load champion name, q, w, e, r abilties
for i in range(0, len(champions)):
    champ_name = champions[i]
    name_len = len(champ_name)
    entities_map = {'entities': []}
    p = ps[i]
    q = qs[i]
    w = ws[i]
    e = es[i]
    r = rs[i]
    
    # get name of each skill and length of the name
    passive_name = p.split(':')[0]
    p_len = len(passive_name)
    q_name = q.split(':')[0]
    q_len = len(q_name)
    w_name = w.split(':')[0]
    w_len = len(w_name)
    e_name = e.split(':')[0]
    e_len = len(e_name)
    r_name = r.split(':')[0]
    r_len = len(r_name)
    
    # find all occurrences of champion's name in the passive
    passive_c_occurrences = [m.start() for m in re.finditer(champ_name, p)]
    for occurrence in passive_c_occurrences:
        entity = (occurrence, occurrence + name_len, 'CHAMP')
        entities_map['entities'].append(entity)
    
    # find all occurrences of champion's passive name in the passive
    passive_a_occurrences = [m.start() for m in re.finditer(passive_name, p)]
    for occurrence in passive_a_occurrences:
        entity = (occurrence, occurrence + p_len, 'ABILITY')
        entities_map['entities'].append(entity)
    
    # place the passive description text and map of all entities in a tuple and append to training_data
    data_entry = (p, entities_map)
    training_data.append(data_entry)
    
    # Same with q
    entities_map = {'entities': []}
    q_c_occurrences = [m.start() for m in re.finditer(champ_name, q)]
    for occurrence in q_c_occurrences:
        entity = (occurrence, occurrence + name_len, 'CHAMP')
        entities_map['entities'].append(entity)
    
    q_a_occurrences = [m.start() for m in re.finditer(q_name, q)]
    for occurrence in passive_a_occurrences:
        entity = (occurrence, occurrence + q_len, 'ABILITY')
        entities_map['entities'].append(entity)
    
    data_entry = (q, entities_map)
    training_data.append(data_entry)
    
    # Same with w
    entities_map = {'entities': []}
    w_c_occurrences = [m.start() for m in re.finditer(champ_name, w)]
    for occurrence in w_c_occurrences:
        entity = (occurrence, occurrence + name_len, 'CHAMP')
        entities_map['entities'].append(entity)
    
    w_a_occurrences = [m.start() for m in re.finditer(w_name, w)]
    for occurrence in w_a_occurrences:
        entity = (occurrence, occurrence + w_len, 'ABILITY')
        entities_map['entities'].append(entity)
    
    data_entry = (w, entities_map)
    training_data.append(data_entry)

    # Same with e
    entities_map = {'entities': []}
    e_c_occurrences = [m.start() for m in re.finditer(champ_name, e)]
    for occurrence in e_c_occurrences:
        entity = (occurrence, occurrence + name_len, 'CHAMP')
        entities_map['entities'].append(entity)
    
    e_a_occurrences = [m.start() for m in re.finditer(e_name, e)]
    for occurrence in e_a_occurrences:
        entity = (occurrence, occurrence + e_len, 'ABILITY')
        entities_map['entities'].append(entity)
    
    data_entry = (e, entities_map)
    training_data.append(data_entry)

    # Same with r
    entities_map = {'entities': []}
    r_c_occurrences = [m.start() for m in re.finditer(champ_name, r)]
    for occurrence in r_c_occurrences:
        entity = (occurrence, occurrence + name_len, 'CHAMP')
        entities_map['entities'].append(entity)
    
    r_a_occurrences = [m.start() for m in re.finditer(r_name, r)]
    for occurrence in r_a_occurrences:
        entity = (occurrence, occurrence + r_len, 'ABILITY')
        entities_map['entities'].append(entity)
    
    data_entry = (r, entities_map)
    training_data.append(data_entry)
```

```python
training_data
```

```python
if 'ner' not in nlp.pipe_names:
    ner = nlp.create_pipe('ner')
    nlp.add_pipe(ner, last=True)
else:
    ner = nlp.get_pipe('ner')
```

```python
ner.add_label("ABILITY")
ner.add_label("CHAMP")
```

```python
for _, annotations in training_data:
    for ent in annotations.get('entities'):
        ner.add_label(ent[2])
    
other_pipe = [pipe for pipe in nlp.pipe_names if pipe != 'ner']
```

```python
import random
from spacy.util import minibatch, compounding
import warnings
from spacy.training.example import Example
```

```python
with nlp.disable_pipes(*other_pipe), warnings.catch_warnings():
    optimizer = nlp.resume_training()
```

```python
for i in range(0, 30):
    print("Starting Iteration" + str(i))
    random.shuffle(training_data)
    losses = {}
    
    for text, annotation in training_data:
        doc = nlp.make_doc(text)
        example = Example.from_dict(doc, annotation)
        nlp.update([example], drop=0.2, sgd=optimizer, losses=losses)

new_model = nlp
```
